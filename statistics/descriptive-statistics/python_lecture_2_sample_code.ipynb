```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_bdd = pd.read_csv('BDD.CSV')
```

```python
df_bdd.info()
```

```python
print(df_bdd['Gender'].value_counts(normalize=True))
print(df_bdd['Disorder'].value_counts(normalize=True))
print(df_bdd['DissLooks'].value_counts(normalize=True))
```

```python
fig, ax = plt.subplots(1,3,figsize=(25,5))
for i, col in enumerate(["Gender","Disorder","DissLooks"]):
  sns.countplot(data=df_bdd,
                x=col,
                ax=ax[i])
```

```python
sns.countplot(data=df_bdd,
                x='Gender',
                hue='DissLooks')
```

```python
sns.countplot(data=df_bdd,
                x='Disorder',
                hue='DissLooks')
```
