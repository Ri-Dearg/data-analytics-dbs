```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df_stroke = pd.read_csv('healthcare-dataset-stroke-data.csv')
```

```python
df_stroke.info()
```

```python
df_stroke.describe()
```

```python
print(df_stroke['gender'].value_counts(normalize=True))
```

```python
fig, ax = plt.subplots(1,3,figsize=(25,5))
for i, col in enumerate(["gender","hypertension","heart_disease"]):
  sns.countplot(data=df_stroke,
                x=col,
                hue='stroke',
                ax=ax[i])
```

```python
fig, ax = plt.subplots(1,3,figsize=(25,5))
for i, col in enumerate(["age","avg_glucose_level","bmi"]):
  sns.histplot(data=df_stroke,
                x=col,
                hue='stroke',
                ax=ax[i])

```

```python
fig, ax = plt.subplots(1,3,figsize=(10,5))
for i, col in enumerate(["age", "avg_glucose_level","bmi"]):
  sns.boxplot(data=df_stroke, x='stroke', y=col,
                ax=ax[i],)

```
